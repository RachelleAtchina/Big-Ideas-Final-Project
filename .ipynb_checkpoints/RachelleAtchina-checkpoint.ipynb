```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fires = pd.read_csv("Fires.csv")
fires.head(10)
```

First, I wanted to sort the different neighborhoods and count how many fires they have had. Some of the data had NaN as their neighborhood, so I excluded that data as it would not be helpful.

```python
counts = fires["neighborhood"].value_counts(dropna=True)
print(counts)
```

Next I filtered based on areas that had less than 40 fires, and got our top 25 best neighborhoods on fires.

```python
filtered_counts = counts[counts < 40]
print(filtered_counts)
```

Next, I wanted to create a graph displaying the top neighborhoods on safety from fires.

```python
filtered_counts.plot(kind = "barh", x= 'fires', y = 'neighborhood', color = 'red')
plt.xlabel("Number of fires")
plt.ylabel("Neighborhoods")
plt.title("Fires In Each Neighborhood")
plt.show()
```

And just because it is visually appealing I wanted to create a map version so we could easily see where the best, safest from a fire, location is in Pittsburgh!

```python
fires_cleaned = fires[['neighborhood', 'latitude', 'longitude']]
fires_cleaned.head()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
#from geodatasets import get_path
```

```python
Longitude = fires['longitude'].tolist()
Latitude = fires['latitude'].tolist()
neighborhoods_map = gpd.read_file("Neighborhoods/Neighborhoods_.shp")
```

```python
fires_cleaned = fires_cleaned.dropna(subset=['longitude', 'latitude'])
gdf = gpd.GeoDataFrame(
    fires_cleaned,
    geometry=gpd.points_from_xy(fires_cleaned['longitude'], fires_cleaned['latitude']),
    crs="EPSG:4326"
)
print(gdf)
```

```python
neighborhoods = gpd.read_file("Neighborhoods/Neighborhoods_.shp") # read in the shapefile
neighborhoods.plot() # plot the geopandas dataframe
plt.show()
plt.clf()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
neighborhoods_map.plot(ax=ax, color="lightgrey", edgecolor="black")
gdf.plot(ax=ax, color="red", markersize=5, label="Fire Locations")

plt.title("Pittsburgh Neighborhoods with Fire Data")
plt.legend()
plt.show()
```

```python
# First, merge the filtered_counts with fires_cleaned
filtered_counts_df = filtered_counts.reset_index(name='Count')
result_df = pd.merge(filtered_counts_df, fires_cleaned, on='neighborhood')

# Create a new column for the "POLYGON" format combining latitude and longitude
result_df['Polygon'] = result_df.apply(
    lambda row: f"POLYGON (({row['longitude']} {row['latitude']}))", axis=1
)

# Display the result
print(result_df[['neighborhood', 'Count', 'Polygon']])
```
